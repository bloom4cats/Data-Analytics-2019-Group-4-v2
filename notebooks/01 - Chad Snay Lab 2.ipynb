Use the [MetaWeather API](https://www.metaweather.com/api/) to retrieve weather data for Columbus on January 1, 2018 using the URL [https://www.metaweather.com/api/location/2383660/2018/1/1]`https://www.metaweather.com/api/location/2383660/2018/1/1` with panda's `read_json()` function; here, the value `2383660` in the HTML link is the [WOEID](https://en.wikipedia.org/wiki/WOEID) for Columbus, Ohio.
2. Display the API documentation from MetaWeather using pandas `read_html` function to display a description of the columns in the resulting DataFrame.
3. Use the `dtypes` function to display the data types of the dataset. Compare the DataFrame results to the API documentation results. Which fields are different? What types of differences might we need to account for in future data cleansing steps?

```python
import pandas as pd

# use pandas to load API data, using API calls to get Columbus, OH, 2018-01-01 data
weather_history = pd.read_json("https://www.metaweather.com/api/location/2383660/2018/1/1")

# use read_html to display the MetaWeather API documentation of each field in the dataset
doc = pd.read_html("https://www.metaweather.com/api")
display(doc[-1][12:24])

# use dtypes to compare data types in the DataFrame to the documentation

## Concatenated API data and DataFrame dtypes
doc[-1]['Compare To API'] = ""    

for index, row in doc[-1].iterrows():
    count = 0
    for weather_entry in weather_history:
        if row[0] == weather_entry:            
            row['Compare To API'] = weather_history.dtypes[count]  
        elif row[0].find('temp') > -1 and weather_entry.find('temp') > -1:           
            row['Compare To API'] = weather_history.dtypes[count]
        doc[-1].loc[index] = row            
        count += 1
        
display(doc[-1][12:24])
    
```
