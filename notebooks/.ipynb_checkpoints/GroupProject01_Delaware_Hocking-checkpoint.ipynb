```python
# Sampling functions
import linecache
import random

def get_line_count(input_file):
    """Count number of lines in a file"""
    count = 0
    with open(input_file) as infile:
        for line in infile:
            count += 1
    return count

def sample_file(input_file, output_file, fraction=0.1):
    """Exctract a subset of lines from a file"""
    total_line_count = get_line_count(input_file)
    sample_line_count = int(fraction * total_line_count)  # fraction of total
    random.seed(12345)  # set an arbitrary number to force random.sample() to return same same sample of rows each time we call sample_file method
    sample_line_numbers = random.sample(range(1, total_line_count), 
                                        sample_line_count)  # sample of line numbers
    sample_line_numbers.sort()
    sample_line_numbers.insert(0, 0)
    with open(output_file, 'w') as outfile:
        for line_number in sample_line_numbers:
            line = linecache.getline(input_file, line_number + 1)
            outfile.write(line)


# Start with Delaware County Cleanup
import pandas as pd

# load full dataset for initial analysis
delaware = pd.read_csv('../data/county_auditor/OH-Delaware/governmaxextract.txt', delimiter="|", low_memory=False)

# sample the data (default is 10% of all rows)
#sample_file('../data/county_auditor/OH-Delaware/governmaxextract.txt', 
#            '../data/county_auditor/OH-Delaware/delaware_county_subset.txt')

#delaware.columns.tolist()

pd.set_option('display.max_column', 200)

delaware_subset = delaware[delaware.mClassificationId >= 500]
delaware_subset = delaware_subset[delaware_subset.mClassificationId <= 599]

delaware_subset = delaware_subset[delaware_subset.SaleAmount < 1500000]
delaware_subset = delaware_subset[delaware_subset.SaleAmount > 0]

delaware_columns = [ 'mpropertyNumber', 'mlegalDescription','mlocStrDir','mlocStrNo','mlocStrNo2',
                     'mlocStrName','mlocStrSuffix','mlocStrSuffixDir','mlocCity','mlocState','mlocZipCode',
                     'mlocDescription','mClassificationId','macres','mfrontfootage','mlot',
                     'msublot','OwnName', 'mtaxsetCode','mstateSchoolcode','MKT_Land_Value','MKT_Impr_Value',
                     'CAUV_Value','MKT_Total_Value','MKT_Tot_Total','SaleDate','SaleAmount',
                     'Conveyance','Hmstd_Flag','LandVal','ImproveVal','SpecAssessment',
                     'AnnualTax','Zoning','Parcel_Type','TresCode']

delaware_subset = delaware_subset[delaware_columns].copy()

address = delaware_subset.mlocStrDir.fillna('') + " " + delaware_subset.mlocStrNo.fillna('') + " " + delaware_subset.mlocStrNo2.fillna('') + " " +delaware_subset.mlocStrName.fillna('') + " " + delaware_subset.mlocStrSuffix.fillna('') + " " + delaware_subset.mlocStrSuffixDir.fillna('')

delaware_subset.insert(2, "Address", address)

# drop the address columns
delaware_subset.drop(['mlocStrDir'], axis=1, inplace=True)
delaware_subset.drop(['mlocStrNo'], axis=1, inplace=True)
delaware_subset.drop(['mlocStrNo2'], axis=1, inplace=True)
delaware_subset.drop(['mlocStrName'], axis=1, inplace=True)
delaware_subset.drop(['mlocStrSuffix'], axis=1, inplace=True)
delaware_subset.drop(['mlocStrSuffixDir'], axis=1, inplace=True)

delaware_subset.rename(
    {'mpropertyNumber' : 'ParcelNumber',
     'mlegalDescription' : 'LegalDescription',
     'Address' : 'Address',
     'mlocCity' : 'City',
     'mlocState' : 'State',
     'mlocZipCode' : 'ZipCode',
     'mlocDescription' : 'Description',
     'mClassificationId' : 'LandUseCode',
     'macres' : 'Acres',
     'mfrontfootage' : 'FrontFootage',
     'mlot' : 'Lot',
     'msublot' : 'SubLot',
     'OwnName' : 'Owner',
     'mtaxsetCode' : 'TaxSetCode',
     'mstateSchoolcode' : 'SchoolDistrict',
     'MKT_Land_Value' : 'MarketLandValue',
     'MKT_Impr_Value' : 'MarketImprovementValue',
     'CAUV_Value' : 'CAUV',
     'MKT_Total_Value' : 'MarketValue',
     'MKT_Tot_Total' : 'MarketTotal',
     'SaleDate' : 'SaleDate',
     'SaleAmount' : 'SalePrice',
     'Conveyance' : 'Conveyance',
     'Hmstd_Flag' : 'HomesteadFlag',
     'LandVal' : 'LandValue',
     'ImproveVal' : 'ImprovementValue',
     'SpecAssessment' : 'SpecAssessment',
     'AnnualTax' : 'AnnualTax',
     'Zoning' : 'Zoning',
     'Parcel_Type' : 'ParcelType',
     'TresCode' : 'TresCode'
    },
    axis=1,
    inplace=True
)

delaware_subset.columns.tolist()
display(delaware_subset.head())
#display(delaware_subset.head())
```
