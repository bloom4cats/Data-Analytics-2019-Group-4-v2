## Selecting Data with Pandas

In this unit we'll continue to work with panda's Series and DataFrames to store and manipulate data.  As part of that work, we'll be interested in examining parts of a larger DataFrame. A common way to select a subset of a DataFrame is through the use of masks and filters - something we've already done, albeit without as much formal discussion about the theories behind using masks and filters. For example, consider the following DataFrame.

<a name="Concept-Masks"></a>
<hr>

#### Concept: DataFrame masks and filters

````python
employees = pd.DataFrame(
    [
        ('John', 'Marketing', '123 Main St', 'Columbus', 'OH', 3),
        ('Jane', 'HR', '456 High Ave', 'Columbus', 'OH', 7),
        ('Bob',  'HR', '152 Market Rd', 'Cleveland', 'OH', 4),
        ('Sue', 'Marketing', '729 Green Blvd', 'Cleveland', 'OH', 8),
        ('Tom', 'IT', '314 Oak Dr', 'Cincinnati', 'OH', 11),
        ('Kate', 'IT', '841 Elm Ln', 'Cincinnati', 'OH', 2)
    ],
    columns = ("name", "department", "address", 
               "city", "state", "years_with_company")
)

display(employees)
````

**Masking**

If we want to work with a specific subset of the data, say only those employees that are in Columbus, we could use the mask `employees.city == 'Columbus'` to filter the data.

````python
employees[employees.city == 'Columbus']
````

**Filtering**

Pandas offers an alternative [`query()` method](https://pandas.pydata.org/pandas-docs/stable/indexing.html#indexing-query) for selecting data that allows us to write statements that are similar to standard Python syntax. For example, `query()` can be used to select only those employees in Columbus.

````python
employees.query('city == "Columbus"')
````

The argument we provide to the `query()` method is a string that indicates the data we'd like to extract from the original DataFame. When comparing string values, we have to to be sure to enclose the inner string in different quotes than the query itself.

Compare the following two lines of code which both return data for employees that have been with the company for more than 4 years.

````python
employees[employees.years_with_company > 4]
````

````python
employees.query('years_with_company > 4')
````

So far there doesn't seem to be much of an advantage to one method over the other. A benefit to using the filtering method `query()` becomes apparent when our conditions become more complex. Compare the two methods when we want to find staff that have been with the company for more than 4 years and live in either Cleveland or Columbus. (NOTE: White space line returns and special indentation has been added here for clarity. It isn't the best Python programming practice to do so in many cases.)

````python
employees[(employees.years_with_company > 4) &
          ((employees.city == "Cleveland") | 
           (employees.city == "Columbus"))]
````

````python
employees.query(
    "years_with_company > 4 and"
    "(city == 'Cleveland' or "
    " city == 'Columbus')"
)
````

The argument supplied to the `query()` method is self-contained in the method call and doesn't require the multiple references to the DataFrame `employees`, making it more readable; however, the `query()` method cannot handle all generic programming situations as gracefully as masking can. For our purposes, we'll tend to stick with the `query()` method for most work we do since it is more readable.

#### Concept: DataFrame masking programmability
One key benefit of DataFrame masking is programmability. As we write scripts, we often use variables to store values that will change, and we can then use the variables in our selection criteria. For example, suppose we have a function that routinely allows different departments to find their staff with more than 4 years with the company.

````python
def senior_staff(df, dept):
    # return only those employees in the 
    # specified department that have been
    # with the company for more than 4 years
    pass    
````

Working with the notation we have been using, writing the function body that returns the correct result for a specified department is straightforward.

````python
def senior_staff(df, dept):
    return df[df.department == dept]   

senior_staff(employees, "IT")
````

<hr>

Working with the `query()` method, there are multiple ways we can achieve the same result.  The method suggested by the pandas' documentation makes use of `@`.  We can reference existing variables in the query string in-line by prefixing their names with `@`. Let's use the `query()` method to search for employees in the HR department:

````python
def senior_staff(df, dept):
    return df.query("department == @dept")

senior_staff(employees, "HR")
````

Let's look at one column to get an idea of what these values represent. Remember that the `saleprice` column represnts the appraised value of the building for each home in our dataset. The rows have the following meanings:

- `count`: the number of non-null elements in the column; here there are 96,560 non-null values in the `saleprice` column
- `mean`: the sum of all values divided by the number of values; the mean value for `saleprice` is 121,249.7 dollars.
- `std`: the standard deviation - a measure of the variation of values within a collection, which can be thought of as an "average" distance to the mean among all the values; the standard deviation of values in the `saleprice` column is 826,601 dollars. A standard deviation this large in comparison to the mean likely indicates a long-tail to our data, which we will explore in more detail shortly.
- `min` and `max`: the smallest and largest values; here we have 0.0 and 27,000,000.00 dollars, respectively. 
- `25%`, `50%`, and `75%`: the quartile values that allow us to divide the data into four parts. The first quartile, 25%, corresponds to the value between the minimum and the median; 25% of values are less than this value.  The second quartile is the median, the middle most number among the values; 50% of values are less than the median and 50% of values are greater than the median. The third quartile represents the middle value between the median and the maximum; 25% of values are greater than this value. 

As we noted previously, `describe()` only display results for numeric columns. For non-numeric columns, we might be interested in knowing about the data values and how often those values appear. Below, we iterate through the columns of the DataFrame and if the the column is a string data we display the column name with the output from the `value_counts()` method.

### Categorical Data

While the examples above provide insight into numeric data, they don't tell us about columns that contain categorical data such as `condition`. 

A typical method of visualizing categorical data is with a bar plot.  Both pandas and Seaborn provide methods of generating bar plots.

For a given column in the DataFrame, the *value_counts()* method returns a Series.  Series, as we've seen before, have a *plot()* method.  We can explicitly create a bar chart using the `kind=bar` or `kind=barh` arguments to the *plot()* method for a vertical or horizontal bar chart, respectively.

````python
df_redux['condition'].value_counts().plot(kind="bar")
````

The order of the categories is determined by the Series used to create the chart; because the results of `value_counts()` are ordered, the bars in the resulting plot are ordered.  

We can create a similar plot using the Seaborn [`countplot()`](https://seaborn.pydata.org/generated/seaborn.countplot.html) function. To create a vertical bar chart, we can use the `x` keyword argument to specify source data; to create a horizontal bar chart, we can use the `y` keyword argument to specify the source data. Below, we create a horizontal bar chart for the count of `condition` data.

````python
sns.countplot(y=df_redux['condition'])
````

Notice that the data isn't ordered from greatest to smallest counts; by default, categories are ordered based on when they appear in the source data, or the natural ordering of the values as numbers. To impose a different ordering, we can use the `order` keyword argument with the `countplot()` function.

Notice also that the values appear to be a single character for most of our data, but written out for other bits of data. We can investigate the types of condition counts by another variable using the `group_by()` function, combined with the `size()` method.

````python
df_redux.groupby(["county", "condition"]).size()
````

Remember that we had to create null value data for Fairfield County since no data existed concerning the condition of the property from the Fairfield county auditor data. Therefore, we can see that Franklin and Licking counties have simply coded their property conditions differently. Let's make them the same and re-run the countplot.

As an alternative to how often a given condition appears in the data, we might be interested in knowing the mean appraised building, appraised land, and sale prices for each dwelling by the condition listed for each dwelling. In order to this, we can use panda's [`group by` functionality](https://pandas.pydata.org/pandas-docs/stable/groupby.html) that allows us to creates groups of data within the dataset and calculate an aggregate value for each group; this is similar to the standard SQL [`group by`](https://www.w3schools.com/sql/sql_groupby.asp) statement.  

To create a grouping, we can use the DataFrame's [`groupby()`](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.groupby.html) method specifying at least one column whose values should be used for grouping. The code below groups the data in `df_redux` by values in the `condition` column, and then computes the mean across the other columns for each group.

````python
df_redux.groupby(['condition']).mean()
````

If we'd like the aggregated values for a single column, we can specify that column after the call to the the `groupby()` method or after the call to the aggregation function. It's generally better to reduce the size of the dataset over which a calculation is applied so we should select the column of interest before doing the aggregation calculation. We can select the `condition` column after grouping the data and then calculate the mean.

````python
mean_saleprice_by_condition = df_redux.groupby(['condition'])['saleprice'].mean()
display(mean_saleprice_by_condition)
````

We can use this aggregated data, stored in a pandas Series, to create a bar chart. Because we are no longer interested in the number of times a `condition` value appears, we cannot use the Seaborn `countplot()` function.  Instead, we can use `barplot()`. Before plotting, we sort the Series using the `sort_values()` method to indicate that we'd like to sort the Series by its values; we specify `inplace=True` to alter the existing Series object and `ascending=False` to sort the values in descending order.

````python
mean_saleprice_by_condition.sort_values(inplace=True, ascending=False)
sns.barplot(x=mean_saleprice_by_condition.index, y=mean_saleprice_by_condition.values)
````

Alternatively, we could have used the `plot()` method on the Series itself.

````python
mean_saleprice_by_condition.plot(kind='barh')
````

### Box Plots

We'll look at modeling these relationships between variables in the next unit. So far, we've used the *describe()* method, histograms, pair plots, and bar charts to get a sense of the distribution and other properties of some of the numeric data in our dataset. Another common way to explore numeric data is through the use of [box plots](https://en.wikipedia.org/wiki/Box_plot).

To understand the components of a box plot, consider the following example Series.

````python
example = pd.Series([-4, -1, -0.5, 0, 0.5, 1, 4], name="Example")
example.describe()
````

Let's create a the box plot for the same Series and compare it to the output above.

````python
example.plot(kind='box')
````

The midline in the box corresponds to the median or 50th percentile value. The lines above and below the median correspond to the 75th and 25th percentile values, respectively. The narrower lines above and below these lines correspond to the maximum and minimum values excluding outliers. Outliers are symbolized by the small circular markers. Outliers represent values that can be considered "distant" from other values. There are various ways of defining what constitutes an outlier but a common method relies on interquartile range, *IQR*, the difference between the third and first quartile, or in terms of percentiles, the difference between the 75th and 25th percentiles. Using interquartile range a value can be considered an outlier if it satisfies one of the following:

- it is greater than the sum of the 75th percentile and 1.5 times the interquartile range 
- it is less than the difference of 25th percentile and 1.5 times the interquartile range.

````python
df_redux.yearbuilt.plot(kind='box')
````

We can see that the median of `yearbuilt` is about 1975, but there are many outliers. This plot begins to illustrate one of the problems with boxplots which should be carefully considered whenever presenting data with this type of chart - the hiding of information. Your reading for this lesson included some discussion of this type of problem, and here we will show how to alleviate it with Python code. First, let's look at two more data columns with even more extreme data hiding concerns.

Let's look at the `bedrooms` and `saleprice` columns.

````python
df_redux.bedrooms.plot(kind='box')
````

The box plot for `bedrooms` is a bit unusual. If we inspect the chart closely we see that the median and the 25th percentile values are the same. This is due to the fact that the distrubition of values are more categorical than continuous, and that  the distribution of those values as shown by `describe()`, `value_counts()`, `skew()`, and `kurtosis()` below reveal a non-normal curve.

````python
display(df_redux.bedrooms.describe())
display(df_redux.bedrooms.value_counts())
display(df_redux.bedrooms.skew())
display(df_redux.bedrooms.kurtosis())
````

If we show the boxplot for `saleprice` we will see an even more extreme example of a boxplot.

````python
df_redux.saleprice.plot(kind='box')
````

Finally, let's consider a hypothetical example of two boxplots which represent two very different data distributions. 

````python
import numpy as np
ex1 = np.random.normal(loc=0, scale=4, size=10000)
example1 = pd.Series(ex1)

ex2 = pd.Series(np.random.normal(loc=-4, scale=1.7, size=5000))
ex3 = pd.Series(np.random.normal(loc=4, scale=1.7, size=5000))
example2 = ex2.append(ex3)

example1.plot(kind='box')
````

````python
example1.plot(kind='hist')
````

````python
example2.plot(kind='box')
````

````python
example2.plot(kind='hist')
````

While the two boxplots look roughly the same (you might think the outliers are not significant in the box plot), the underlying data structures are considerably different. In example 1, the data is a normal distribution, but in example 2 the data is a bi-nomial distribution. Therefore, when using boxplots, either for analysis or display, be sure to analyze the data for the appropriate type of homogeneity or heterogeneity that needs to exist in order to make the boxplot a meaningful representation of the results of sampling, testing, or experimental design responses. Failing to do so could cause later efforts to develop a predictive model difficult to achieve.

Let's calculate thresholds for the outliers of `area`.

````python
q1 = df_redux.area.quantile(0.25)  # first quartile
q3 = df_redux.area.quantile(0.75)  # third quartile
IQR = q3 - q1  # IQR
lower_threshold = q1 - 1.5 * IQR  # lower threshold
upper_threshold = q3 + 1.5 * IQR  # upper threshold
display(lower_threshold, upper_threshold)
````

From these calculations, an outlier for the area of a dwelling is anything greater than about 3400 sq. ft., or less than -213 sq. ft.; therefore, for this data there are no lower outliers since negative values are meaningless.

### Pivot Tables

Pandas provides the ability to create [pivot tables](https://en.wikipedia.org/wiki/Pivot_table) to aggregate and summarize data.  To crate a pivot table, we can use the DataFrame's [*pivot_table()*](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.pivot_table.html) method and, at a minimum, specifying which column should be used as the index of the new table; the index serves as the column by which values are grouped and aggregated.  By default, the mean is used as the aggregation function but we can specify any appropriate function using the `aggfunc` keyword argument.

Let's create a pivot table that groups the data by `year` and aggregates data by calculating the median of values.

````python
df_redux.pivot_table(index="condition", aggfunc=pd.np.median)
````

We can choose a subset of columns by listing them using the `values` keyword argument.

````python
df_redux.pivot_table(index="condition", values=["saleprice", "area", "appraisedtaxablebuilding"], aggfunc=pd.np.median)
````

We fan also further divide the data by specifying additional indexes or through the `columns` keyword argument.

````python
df_redux.pivot_table(index=["county", "condition"], values=["saleprice", "area", "appraisedtaxablebuilding"], aggfunc=pd.np.median)
````

````python
df_redux.pivot_table(index="county", 
                       columns="condition",
                       values=["saleprice", "area", "appraisedtaxablebuilding"], 
                       aggfunc=pd.np.median)
````

We an also use pivot tables as inputs for visualizations. Let's compare sale price and number of bedrooms.

````python
pivot = (
    df_redux.pivot_table(
        index=["condition", "bedrooms"],
        values=["saleprice"],
        aggfunc=pd.np.median)
    .sort_values(by="saleprice", ascending=False))
display(pivot)
````

It is interesting here that the sale price is lowest for dwellings with 10 bedrooms. Perhaps these dwellings represent a single structure with multiple condominiums or apartments contained within them? In any case, with this pivot table we can create a [heatmap](https://en.wikipedia.org/wiki/Heat_map) using Seaborn's [*heatmap()*](https://seaborn.pydata.org/generated/seaborn.heatmap.html) function. In addition to specifying the pivot table as the data source, we can show the values for each rectangle using the `annot` keyword argument and specify the colormap using the `cmap` keyword argument with a [matplotlib colormap  name](https://matplotlib.org/users/colormaps.html).

````python
sns.heatmap(pivot, annot=True, cmap="RdYlGn")
````

So we might infer from the heatmap that dwellings with a `condition` rating of "Good" or "Very Good" tend to rate much higher than "Fair" or "Poor" condition dwellings which is what we would expect. Additionally, we can infer that bedroom counts do not seem to be a particularly good predictor of sale price, which could be caused by multiple other factors that we are not looking at in this chart. This result doesn't necessarily tell us whether more bedrooms will increase the value of any one dwelling more or less than another, and it doesn't specify what any one individual's decision-making process should be for any single home. However, it does give us some insight into how these two variables may or may not correlate to one another, which may help us when we begin modeling our data in the next Unit.

### Time-Series Plots

When dealing with dataframes containing historical data it is useful to view metrics for trends over time. If you've ever used Microsoft Excel you've almost certainly built a simple line chart for a time series variable. In data analytics, we are not just concerned with viewing the time-series to talk about the latest value of the stock market, business revenues, or profits compared to previous levels. Primarily we are looking for trends in the data which we can model or consider in our analysis of the dataset. Let's return to our Realtor.com historical data to illustrate time-series plots.

````python
#from sqlalchemy import create_engine
#engine = create_engine('sqlite:///../data/output.sqlite')

query = "SELECT * from realtor_history;"
realtor_history = pd.read_sql_query(query, con=engine)
display(realtor_history.head())
````

Remember that this dataset contains data by `month` and by `county`. We can use these two variables to build a line plot.

````python
from matplotlib import pyplot as plt

franklin = realtor_history[realtor_history['county'] == 'Franklin']
franklin['month'] = pd.to_datetime(franklin['month'], format='%Y-%m-%d', errors='coerce')
#franklin.sort_values('month', ascending=True, inplace=True)

plt.plot(franklin['month'], franklin['median_listing_price'], '-')
````

A fairly simple line plot, but not very informative. Let's add a few features to our lineplot to improve readability. Let's also look at the average as well as median listing prices on the same plot.

````python
plt.figure(num=None, figsize=(10, 6))
plt.plot(franklin['month'], franklin['median_listing_price'], color='blue')
plt.plot(franklin['month'], franklin['avg_listing_price'], color='orange')
plt.title('Realtor.com Listing Price')
plt.xlabel('by Month')
plt.ylabel('Listing Price')
plt.legend()
plt.show()
````

Here we see a more readable version of our line plots for both metrics. We can also see some interesting patterns in these charts, and trends that appear to repeat as well as show a steady growth over time - at least for our Franklin county home prices.
